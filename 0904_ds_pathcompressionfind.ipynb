```python
import graphviz as gv
```

```python
def drawDS(ds):
  graph = gv.Digraph("DisjointSet")
  graph.graph_attr['rankdir'] = "BT"
  for e, p in enumerate(ds):
    graph.node(str(e))
    if p >= 0:
      graph.edge(str(e), str(p))
  return graph
```

```python
def find(ds, n):
  if ds[n] >= 0:
    n = ds[n]
  return n
```

```python
def union(ds, a, b):
  a = find(ds,a)
  b = find(ds,b)

  if a == b: return
  
  if ds[a] < ds[b]:
    ds[a] += ds[b]
    ds[b] = a
  else:
    ds[b]+= ds[a]
    ds[a] = b
```

```python
ds= [-1]*10
drawDS(ds)
```

```python
union(ds, 3,4)
drawDS(ds)
```

```python
union(ds, 4,9)
drawDS(ds)
```
