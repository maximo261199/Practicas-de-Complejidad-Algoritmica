```python
with open("1001a.al") as f:
  G = []
  for line in f:
    if line == '-\n':
      G.append([])
    else:
      nums = [int(x) for x in line.split()]
      G.append([(nums[i], nums[i+1]) for i in range(0, len(nums), 2)])

  for x in G:
    print(x)
```

```python
adjlShow(G, weighted=True, layout="neato")
```

```python
import heapq as h
import math

def prim(G):
  n = len(G)
  padre = [-1]*n
  costo = [math.inf]*n
  visitado = [False]*n
  
  costo[0] = 0
  q =[(0,0)]
  while q:
    _, u = h.heappop(q)
    if visitado[u]:continue
    visitado[u] = True
    for v,w in G[u]:
      if not visitado[v] and w < costo[v]:
        costo[v] = w
        padre[v] = u
        h.heappush(q,(w,v))
  return padre
 


```

```python
path = prim(G)
print(path)
adjlShow(G, weighted=True, path=path, layout="neato")
```
