```python
 import graphviz as gv
```

```python
def drawDS(ds):
  graph = gv.Digraph("DisjointSet")
  graph.graph_attr['rankdir'] = "BT"
  for e, p in enumerate(ds):
    graph.node(str(e))
    if p != e:
      graph.edge(str(e), str(p))
  return graph
```

```python
ds = [0, 1, 2, 2, 8, 5, 5, 7, 8, 2]
drawDS(ds)
```

#Quick Union

```python
def find(ds, e):
  while ds[e] != e:
    e = ds[e]
  return e
```

```python

find(ds, 0), find(ds, 7), find(ds, 3), find(ds, 9), find(ds, 2)
```

```python
def union(ds, a, b):
  a = find(ds,  a)
  b = find(ds, b)

  if a == b: return

  ds[a] = b
```

```python
ds1 = [i for i in range(10)]
drawDS(ds1)
```

```python
union(ds1, 3, 4)
drawDS(ds1)
```
